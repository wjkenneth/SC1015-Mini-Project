```python
# Basic Libraries
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
sb.set() # set the default Seaborn style for graphics
```

# Mean years of schooling => 1980 to 2022

```python
schooling_years_raw = pd.read_excel("Raw Data/Mean Years Of Schooling, Annual.xlsx", header = 9)
schooling_years_raw = schooling_years_raw.iloc[:4,:]
schooling_years_raw
```

```python
schooling_years_overall = (schooling_years_raw.iloc[:2,1:].T).iloc[::-1]
#schooling_years_overall.loc["Unnamed: 1", 0] = 5
schooling_years_overall.index = list(range(schooling_years_overall.shape[0]))
#schooling_years_raw
schooling_years_overall.columns = ['Year', 'Mean years of schooling for >25']
schooling_years_overall
```

```python
schooling_years_overall.plot(x = "Year", y = "Mean years of schooling for >25", figsize=(10, 6))
```

```python
schooling_years_overall_all = (schooling_years_raw.iloc[:,1:].T).iloc[::-1]
schooling_years_overall_all.index = list(range(schooling_years_overall_all.shape[0]))
schooling_years_overall_all.columns = ['Year', 'Overall', "Males", "Females"]
schooling_years_overall_all["Year"] = pd.to_datetime(schooling_years_overall_all["Year"].astype(float), format="%Y")
schooling_years_overall_all
#Mean years of schooling for >25
```

```python
import matplotlib.pyplot as plt # we only need pyplot
schooling_years_overall_all.plot(x = "Year", y = ['Overall', "Males", "Females"], figsize=(10, 6))
plt.xlabel('Year')
plt.ylabel('Years (of schooling)')
plt.title("Mean years of schooling for >25")
#plt.xlabel('Year')
#plt.ylabel('Marriage Rate')
#plt.title('Male Marriage Rate')
plt.show()
```

```python
schooling_years_overall_all.to_excel("Cleaned Data/Mean Years Of Schooling.xlsx")
```
