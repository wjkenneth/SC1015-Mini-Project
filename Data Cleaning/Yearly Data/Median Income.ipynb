```python
# Basic Libraries
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
sb.set() # set the default Seaborn style for graphics
```

```python
median_income_overall_raw = pd.read_excel("Raw Data/Gross Monthly Income From Work (Including Employer CPF) Of Full-Time Employed Residents, Annual.xlsx", header=9)
median_income_overall_raw
```

```python
median_data_overall_sieved = median_income_overall_raw.iloc[[0,1],1:]
median_data_overall_sieved
```

```python
median_data_overall_t = median_data_overall_sieved.T.copy()
median_data_overall_t
median_data_overall = median_data_overall_t.iloc[::-1]
median_data_overall.columns = ["Year", "Median Income"]
median_data_overall.index = list(range(median_data_overall.shape[0]))
median_data_overall = median_data_overall.apply(pd.to_numeric, errors="coerce")
median_data_overall_temp = median_data_overall.copy()
median_data_overall["Year"] = pd.to_datetime(median_data_overall["Year"].astype(float), format="%Y")
median_data_overall
```

```python
median_data_overall.plot(x="Year", y="Median Income", figsize=(10, 6))
```

```python
median_data_overall.to_excel("Cleaned Data/Median Income.xlsx")
```
