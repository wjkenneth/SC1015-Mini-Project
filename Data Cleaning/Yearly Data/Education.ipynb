```python
education_female_split_under40_raw = education_attained_raw.iloc[[0, 162, 163, 164, 172, 173, 174],1:]
education_female_split_under40_raw
```

```python
education_female_split_under40_t = education_female_split_under40_raw.T.copy()
for i in range(education_female_split_under40_t.shape[0]):
    education_female_split_under40_t.iloc[i, 1] = education_female_split_under40_t.iloc[i, 1] + education_female_split_under40_t.iloc[i, 2] + education_female_split_under40_t.iloc[i, 3]
    education_female_split_under40_t.iloc[i, 4] = education_female_split_under40_t.iloc[i, 4] + education_female_split_under40_t.iloc[i, 5] + education_female_split_under40_t.iloc[i, 6]
education_female_split_under40 = education_female_split_under40_t[[0, 162, 172]]
education_female_split_under40.columns = ["Year", "Female_Poly 25_39", "Female_Uni 25_39"]
education_female_split_under40 = education_female_split_under40.iloc[::-1]
education_female_split_under40.index = list(range(education_female_split_under40.shape[0]))
education_female_split_under40["Year"] = pd.to_datetime(education_female_split_under40["Year"], format="%Y")
education_female_split_under40
```

```python
##plt.figure(figsize=(16, 8), dpi=150)

#education_under35_level_split_raw["Poly 25_34"].plot(label = 'Poly', color = 'r')
#education_under35_level_split_raw["Uni 25_34"].plot(label = 'Uni', color = 'b')
#education_under35_level_split_raw["Year"].plot(xlabel)

##education_female_split_under40.plot(color =['r','b'])
##plt.title("Education level at 25 to 39 for female, so number of years spent studying, harder to get a child involved")
##plt.xlabel('Years')
##plt.xticks(education_female_split_under40.index, education_female_split_under40["Year"].values)
##plt.legend()

education_female_split_under40.plot(x = "Year", y =["Female_Poly 25_39", "Female_Uni 25_39"], figsize=(10, 6))
```
