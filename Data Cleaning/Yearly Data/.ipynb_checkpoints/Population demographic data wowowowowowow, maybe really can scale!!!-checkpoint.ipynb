```python
# some headstart on dependency!!
dependency_raw = ((key_annual_overall.iloc[[0, 24, 27],:].T).copy()).iloc[::-1]
dependency_raw.columns = dependency_raw.iloc[-1]
dependency_raw = dependency_raw.drop(dependency_raw.index[-1])
#dependency_raw.index = list(range(dependency_raw.shape[0]))
#dependency_raw.index.name = None
dependency_raw = dependency_raw.rename_axis(None, axis=1)
dependency_raw = dependency_raw.reset_index().drop("index", axis=1)
dependency_raw = dependency_raw.rename(columns={"Data Series":"Year"})
dependency_raw = dependency_raw.apply(pd.to_numeric, errors="coerce")
dependency_raw["Year"] = pd.to_datetime(dependency_raw["Year"], format="%Y")
dependency_raw.head(50)
```

```python
## maybe extrapolate the middle na values... or...
dependency_from1990 = dependency_raw.iloc[40:,:].copy()
dependency_from1990 = dependency_from1990.reset_index().drop("index", axis=1)
dependency_from1990
```

```python
sg_residents_only = pd.read_excel("Raw Data/Singapore Residents By Age Group, Ethnic Group And Sex, End June, Annual.xlsx", header=9)
sg_residents_only.head(10)
```

# total residents is mentioned above already!!, but here is 1957 so better?

```python
sg_residents_only_overall = ((sg_residents_only.iloc[:26,1:]).T).iloc[::-1]
sg_residents_only_overall.index = list(range(sg_residents_only_overall.shape[0]))
sg_residents_only_overall[0] = pd.to_datetime(sg_residents_only_overall[0].astype(float), format="%Y")
sg_residents_only_overall
```

```python
sg_residents_total = sg_residents_only_overall.iloc[:,:2].copy()
sg_residents_total.columns = ["Year", "Total Residents"]
#sg_residents_total["Year"] = pd.to_datetime(sg_residents_total["Year"].astype(float), format="%Y")
sg_residents_total
```

```python
sg_residents_total.plot(x="Year", y="Total Residents", figsize=(10, 6))
```

```python
sg_residents_within_age_under39 = (sg_residents_only_overall.iloc[:,:10]).copy() ## above 20 too ah, 20-39
for i in range(sg_residents_within_age_under39.shape[0]):
    sg_residents_within_age_under39.iloc[i, 1] = sg_residents_within_age_under39.iloc[i, 6] + sg_residents_within_age_under39.iloc[i, 7] + sg_residents_within_age_under39.iloc[i, 8] + sg_residents_within_age_under39.iloc[i, 9]
sg_residents_within_age_under39 = sg_residents_within_age_under39.iloc[:,:2]
sg_residents_within_age_under39.columns = ["Year", "Number of 20-39 RESIDENTS"]
sg_residents_within_age_under39
```

```python
sg_residents_within_age_under39.plot(x="Year", y="Number of 20-39 RESIDENTS", figsize=(10,6))
```

# since females give birth and their organs die out earlier sort of, so females 20-39 to see also!

```python
sg_residents_female_within_age_under39 = ((sg_residents_only.T).iloc[1:,[0,56,57,58,59]]).iloc[::-1]
for i in range(sg_residents_female_within_age_under39.shape[0]):
    sg_residents_female_within_age_under39.iloc[i,1] = sg_residents_female_within_age_under39.iloc[i,1] + sg_residents_female_within_age_under39.iloc[i,2] + sg_residents_female_within_age_under39.iloc[i,3] + sg_residents_female_within_age_under39.iloc[i,4]
sg_residents_female_within_age_under39 = sg_residents_female_within_age_under39.iloc[:,:2]
sg_residents_female_within_age_under39.index = list(range(sg_residents_female_within_age_under39.shape[0]))
sg_residents_female_within_age_under39.columns = ["Year", "Number of Female 20-39 RESIDENTS"]
sg_residents_female_within_age_under39["Year"] = pd.to_datetime(sg_residents_female_within_age_under39["Year"].astype(float), format="%Y")
sg_residents_female_within_age_under39
```

```python
sg_residents_female_within_age_under39.plot(x="Year", y="Number of Female 20-39 RESIDENTS", figsize=(10, 6))
```

```python
all_together_sg_residents = sg_residents_total.copy()
all_together_sg_residents = all_together_sg_residents.merge(sg_residents_within_age_under39)
all_together_sg_residents = all_together_sg_residents.merge(sg_residents_female_within_age_under39)
all_together_sg_residents
```

```python
all_together_sg_residents.plot(x="Year", y=["Total Residents", "Number of 20-39 RESIDENTS", "Number of Female 20-39 RESIDENTS"], figsize=(10, 6))
```
