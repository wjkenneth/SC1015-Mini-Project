### Import Libraries

```python
# Basic Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Import Data

```python
df = pd.read_excel('Raw Data/Marriage Rates, Annual.xlsx', header = None, dtype=str)
```

### Data Cleaning

```python
# Remove rows with NaN values
df = df.dropna()

# Transpose the DataFrame
df = df.T

# Use the first row as the header
df.columns = df.iloc[0]

# Drop the first row
df = df[1:]

df.head()
```

### Extracting Columns

```python
# select columns
marriage_rate = df[['Data Series', 'Male General Marriage Rate (Per 1,000 Unmarried Resident Males/Females Aged 15-49 Years)','Female General Marriage Rate (Per 1,000 Unmarried Resident Males/Females Aged 15-49 Years)','Crude Marriage Rate (Per 1,000 Residents)']]

# sort dataframe by year column
marriage_rate = marriage_rate.sort_values(by='Data Series')

marriage_rate = marriage_rate.rename(columns={'Data Series': 'Year', 
                                              'Male General Marriage Rate (Per 1,000 Unmarried Resident Males/Females Aged 15-49 Years)': 'Male Marriage Rate',
                                              'Female General Marriage Rate (Per 1,000 Unmarried Resident Males/Females Aged 15-49 Years)' : 'Female Marriage Rate', 
                                              'Crude Marriage Rate (Per 1,000 Residents)' : 'Crude Marriage Rate'})

# strip blank space
marriage_rate['Year'] = marriage_rate['Year'].str.strip()

# convert the "Marriage Rate" column to float format
marriage_rate['Male Marriage Rate'] = marriage_rate['Male Marriage Rate'].astype(float)
marriage_rate['Female Marriage Rate'] = marriage_rate['Female Marriage Rate'].astype(float)
marriage_rate['Crude Marriage Rate'] = marriage_rate['Crude Marriage Rate'].astype(float)

# convert the "Year" column to datetime format
marriage_rate['Year'] = pd.to_datetime(marriage_rate['Year'], format='%Y')
```

### Plot Yearly Data

```python
# Plot the data
marriage_rate.plot(x='Year', y=['Male Marriage Rate','Female Marriage Rate','Crude Marriage Rate'], figsize=(10, 6))

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Marriage Rate')
plt.title('Male Marriage Rate')

# Show the plot
plt.show()
```

### Exporting Data

```python
# Export yearly data to CSV
marriage_rate.to_csv('Cleaned Data/Marriage Rate.csv')
```
