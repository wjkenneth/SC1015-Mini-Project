```python
# Basic Libraries
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
sb.set() # set the default Seaborn style for graphics
```

# Income Data, median, => 1996 to 2022, missing year 2000 and 2005 data somehow from MOM...

```python
median_income_overall_raw = pd.read_excel("Raw Data/Gross Monthly Income From Work (Including Employer CPF) Of Full-Time Employed Residents, Annual.xlsx", header=9)
median_income_overall_raw
```

```python
median_data_overall_sieved = median_income_overall_raw.iloc[[0,1],1:]
#median_data_overall_sieved.loc[0, "Dollar"] = 55555
median_data_overall_sieved
```

```python
median_data_overall_t = median_data_overall_sieved.T.copy()
median_data_overall_t
median_data_overall = median_data_overall_t.iloc[::-1]
median_data_overall.columns = ["Year", "Median Income"]
median_data_overall.index = list(range(median_data_overall.shape[0]))
median_data_overall = median_data_overall.apply(pd.to_numeric, errors="coerce")
median_data_overall_temp = median_data_overall.copy()
median_data_overall["Year"] = pd.to_datetime(median_data_overall["Year"].astype(float), format="%Y")
median_data_overall
```

```python
median_data_overall.plot(x="Year", y="Median Income", figsize=(10, 6))
```

### income change per month? extapolate but not very good, very naive

## Better to just stick to annual honestly, scale to monthly very weird for some data...

### just keep same ah, for 12 months and change every 12

```python
median_data_overall_monthly = pd.DataFrame(columns=["Year_month", "Median Income"])
for i in range(median_data_overall_temp.shape[0]*12):
    median_data_overall_monthly.loc[i] = [f"{int(median_data_overall_temp.iloc[i//12, 0])}/{(i%12)+1}", median_data_overall_temp.iloc[i//12, 1]]
median_data_overall_monthly["Year_month"] = pd.to_datetime(median_data_overall_monthly["Year_month"], format="%Y %m")
median_data_overall_monthly
```

```python
median_data_overall_monthly.plot(x="Year_month",y="Median Income", figsize=(10, 6))
```

```python
df = median_data_overall_monthly.copy()
#for i in range(df.shape[0]):
#    df.loc
df["Median Income"] = np.NaN
for i in range(median_data_overall.shape[0]):
    df.loc[4+(i*12), "Median Income"] = median_data_overall.iloc[i,1] ## 5th month ah just like that...
df.tail(10)
df
```

```python
from scipy.interpolate import CubicSpline

# convert the 'Year' column to datetime format
#df['Year'] = pd.to_datetime(df['Year'])

# set the 'Year' column as the index
#df = df.set_index('Year_month')

# create a cubic spline interpolation function
cs = CubicSpline(df.dropna().index, df.dropna()['Median Income'])

# use the interpolation function to fill in the missing values
df['Median Income'] = cs(df.index)

# reset the index
#df = df.reset_index()

# print the resulting data frame
print(type(df))
df
```

```python
df.plot(x="Year_month", y="Median Income", figsize=(10, 6))
```
