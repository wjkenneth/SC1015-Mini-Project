```python
# Basic Libraries
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
sb.set() # set the default Seaborn style for graphics
```

```python
annual_live_birth = pd.read_excel("Cleaned Data/Live Births.xlsx", usecols=lambda x: 'Unnamed' not in x)
annual_total_population = pd.read_excel("Cleaned Data/Total Population.xlsx", usecols=lambda x: 'Unnamed' not in x)
annual_sg_residents = pd.read_excel("Cleaned Data/Singapore Population.xlsx", usecols=lambda x: 'Unnamed' not in x)
annual_brides_under39 = pd.read_excel("Cleaned Data/Marriages under 39.xlsx", usecols=lambda x: 'Unnamed' not in x)
annual_brides_under44 = pd.read_excel("Cleaned Data/Marriages under 44.xlsx", usecols=lambda x: 'Unnamed' not in x)
annual_marriages_overall = pd.read_excel("Cleaned Data/Total Marriages.xlsx", usecols=lambda x: 'Unnamed' not in x)
annual_schooling_years = pd.read_excel("Cleaned Data/Mean Years Of Schooling.xlsx", usecols=lambda x: 'Unnamed' not in x)
annual_old_young_dependency = pd.read_excel("Cleaned Data/Old Age Dependency.xlsx", usecols=lambda x: 'Unnamed' not in x)
annual_highest_education_overall = pd.read_excel("Cleaned Data/Highest Education Attained.xlsx", usecols=lambda x: 'Unnamed' not in x)
annual_female_poly_uni_under40_tgt = pd.read_excel("Cleaned Data/Females Under 40 Completed Poly&Uni.xlsx", usecols=lambda x: 'Unnamed' not in x)
annual_poly_uni_under_35_tgt = pd.read_excel("Cleaned Data/Completed Poly&Uni.xlsx", usecols=lambda x: 'Unnamed' not in x)
annual_median_income = pd.read_excel("Cleaned Data/Median Income.xlsx", usecols=lambda x: 'Unnamed' not in x)
annual_household_monthly_median_income = pd.read_csv("Cleaned Data/Median Monthly Household Income.csv")
annual_avg_hours_worked = pd.read_csv("Cleaned Data/Average Working Hours.csv")
annual_fertility_rate = pd.read_csv("Cleaned Data/Fertility Rate.csv")
annual_marriage_rate = pd.read_csv("Cleaned Data/Marriage Rate.csv")
annual_unemployment_rate = pd.read_csv("Cleaned Data/Unemployment Rate.csv")
```

```python
annual_combined_data = annual_live_birth.copy()
annual_combined_data = annual_combined_data.merge(annual_total_population, on="Year", how="left")
annual_combined_data = annual_combined_data.merge(annual_sg_residents, on="Year", how="left")
annual_combined_data = annual_combined_data.merge(annual_brides_under39, on="Year", how="left")
annual_combined_data = annual_combined_data.merge(annual_brides_under44, on="Year", how="left")
annual_combined_data = annual_combined_data.merge(annual_marriages_overall, on="Year", how="left")
annual_combined_data = annual_combined_data.merge(annual_schooling_years, on="Year", how="left")
annual_combined_data = annual_combined_data.merge(annual_old_young_dependency, on="Year", how="left")
annual_combined_data = annual_combined_data.merge(annual_highest_education_overall, on="Year", how="left")
annual_combined_data = annual_combined_data.merge(annual_female_poly_uni_under40_tgt, on="Year", how="left")
annual_combined_data = annual_combined_data.merge(annual_poly_uni_under_35_tgt, on="Year", how="left")
annual_combined_data = annual_combined_data.merge(annual_median_income, on="Year", how="left")
annual_combined_data = annual_combined_data.merge(annual_household_monthly_median_income, on="Year", how="left")
annual_combined_data = annual_combined_data.merge(annual_avg_hours_worked, on="Year", how="left")
annual_combined_data = annual_combined_data.merge(annual_fertility_rate, on="Year", how="left")
annual_combined_data = annual_combined_data.merge(annual_marriage_rate, on="Year", how="left")
annual_combined_data = annual_combined_data.merge(annual_unemployment_rate, on="Year", how="left")
annual_combined_data
```

```python
annual_combined_data.to_excel("Cleaned Data/Yearly Data.xlsx")
```
