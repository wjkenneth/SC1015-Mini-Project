```python
all_marriage_data_raw = pd.read_excel("Raw Data/Key Indicators On Marriages, Annual.xlsx", header = 8)
all_marriage_data_raw
```

```python
brides_age_data_marriage_raw = pd.read_excel("Raw Data/Marriages Under The Women's Charter By Age Group Of Brides, Annual.xlsx", header = 9)
brides_age_data_marriage_raw
# useful maybe more cos their age to give birth and get pregnant, fertility and all!!
```

```python
marriages_total_overall = (all_marriage_data_raw.iloc[:2,1:].T).iloc[::-1]
marriages_total_overall.columns = ["Year", "Number (Total)"]
marriages_total_overall.index = list(range(marriages_total_overall.shape[0]))
marriages_total_overall["Year"] = pd.to_datetime(marriages_total_overall["Year"].astype(float), format="%Y")
marriages_total_overall
```

```python
marriages_total_overall.plot(x="Year", y="Number (Total)", figsize=(10, 6))
plt.title("Number of Marriages, each year")
plt.show()
```

```python
marriages_brideside_overall = (brides_age_data_marriage_raw.iloc[:2,1:].T).iloc[::-1].copy()
marriages_brideside_overall.columns = ["Year", "Number (Total)"]
marriages_brideside_overall.index = list(range(marriages_total_overall.shape[0]))
marriages_brideside_overall["Year"] = pd.to_datetime(marriages_brideside_overall["Year"].astype(float), format="%Y")
marriages_brideside_overall
```

```python
marriages_brideside_overall.plot(x="Year", y="Number (Total)", figsize=(10, 6))
plt.title("Number of Marriages, each year")
plt.show()
```

```python
marriages_under39_overall = ((brides_age_data_marriage_raw.iloc[:7,1:].T).iloc[::-1]).drop(1, axis=1)
#marriages_under39_overall.columns = ["Year", "Number (Total)"]
marriages_under39_overall.index = list(range(marriages_under39_overall.shape[0]))
for i in range(marriages_under39_overall.shape[0]):
    marriages_under39_overall.loc[i, 2] = marriages_under39_overall.iloc[i,1] + marriages_under39_overall.iloc[i,2] + marriages_under39_overall.iloc[i,3] + marriages_under39_overall.iloc[i,4] + marriages_under39_overall.iloc[i,5]
marriages_under39_overall = marriages_under39_overall.iloc[:,:2]
marriages_under39_overall.columns = ["Year", "Number (Total under 39)"]
marriages_under39_overall["Year"] = pd.to_datetime(marriages_under39_overall["Year"].astype(float), format="%Y")
marriages_under39_overall
```

```python
marriages_under39_overall.plot(x="Year", y="Number (Total under 39)", figsize=(10, 6))
```

```python
#marriages_under44_overall - all_marriage_data_raw
marriages_under44_overall = ((brides_age_data_marriage_raw.iloc[:8,1:].T).iloc[::-1]).drop(1, axis=1)
#marriages_under39_overall.columns = ["Year", "Number (Total)"]
marriages_under44_overall.index = list(range(marriages_under44_overall.shape[0]))
for i in range(marriages_under44_overall.shape[0]):
    marriages_under44_overall.loc[i, 2] = marriages_under44_overall.iloc[i,1] + marriages_under44_overall.iloc[i,2] + marriages_under44_overall.iloc[i,3] + marriages_under44_overall.iloc[i,4] + marriages_under44_overall.iloc[i,5] + marriages_under44_overall.iloc[i,6]
marriages_under44_overall = marriages_under44_overall.iloc[:,:2]
marriages_under44_overall.columns = ["Year", "Number (Total under 44)"]
marriages_under44_overall["Year"] = pd.to_datetime(marriages_under44_overall["Year"].astype(float), format="%Y")
marriages_under44_overall
```

```python
marriages_under44_overall.plot(x="Year", y="Number (Total under 44)", figsize=(10, 6))
```
