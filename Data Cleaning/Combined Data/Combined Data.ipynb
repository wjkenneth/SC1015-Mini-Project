```python
# Basic Libraries
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
sb.set() # set the default Seaborn style for graphics
```

```python
annual_live_birth = pd.read_excel("annual 1960 to 2022 live birth data excel.xlsx", usecols=lambda x: 'Unnamed' not in x)

annual_inflation = pd.read_excel("annual(from monthly) 1962 to 2022 inflation data excel.xlsx", usecols=lambda x: 'Unnamed' not in x)
annual_cpi = pd.read_excel("annual(from monthly) 1962 to 2022 CPI data excel.xlsx", usecols=lambda x: 'Unnamed' not in x)
annual_marriage = pd.read_excel("annual(from monthly) 1961 to 2022 marriage number data excel.xlsx", usecols=lambda x: 'Unnamed' not in x)
annual_infant_death = pd.read_excel("annual(from monthly) 1960 to 2022 infant death number data excel.xlsx", usecols=lambda x: 'Unnamed' not in x)
annual_average_houses_soldprice = pd.read_excel("annual(from monthly) monthly_average_houses_price.xlsx", usecols=lambda x: 'Unnamed' not in x)
annual_house_sold_count = pd.read_excel("annual(from monthly)_count_houses_sold.xlsx", usecols=lambda x: 'Unnamed' not in x)
```
