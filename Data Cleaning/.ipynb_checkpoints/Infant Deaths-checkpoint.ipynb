```python
infant_death_no = pd.read_excel("Raw Data/Infant Deaths By Ethnic Group And Sex, Monthly.xlsx", header = 9)
infant_death_no.head()
```

```python
infant_death_no = infant_death_no.iloc[[0,1],1:]
infant_death_no.head()
```

```python
infant_death_no_t = infant_death_no.T.iloc[::-1].copy()
infant_death_no_t.columns = ["Data Series_monthly", "Total Infant Deaths"]
infant_death_no_t.index = list(range(756)) ##inflation_rate_t.index = list(range(733))
infant_death_no_t
```

```python
##inflation_rate_t['year_Month'] = pd.to_datetime(inflation_rate_t['year_Month'])
infant_death_no_t["Data Series_monthly"] = pd.to_datetime(infant_death_no_t["Data Series_monthly"])
infant_death_no_t
```

```python
## inflation_rate_t.plot(x='year_Month', y='Inflation')
infant_death_no_t.plot(x = "Data Series_monthly", y = "Total Infant Deaths", figsize=(10, 6))
```

```python
infant_death_no_t.to_excel("Cleaned Data/No. of Infant Deaths.xlsx")
```
