```python
# Basic Libraries
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
sb.set() # set the default Seaborn style for graphics
```

```python
monthly_live_birth = pd.read_excel("Cleaned Data/Live Births.xlsx", usecols=lambda x: 'Unnamed' not in x)
monthly_infant_death = pd.read_excel("Cleaned Data/Infant Deaths.xlsx", usecols=lambda x: 'Unnamed' not in x)
monthly_marriage_number = pd.read_excel("Cleaned Data/No. of Marriages, Monthly.xlsx", usecols=lambda x: 'Unnamed' not in x)
monthly_CPI_seasonally_adjusted = pd.read_excel("Cleaned Data/CPI, Monthly, Seasonally Adjusted.xlsx", usecols=lambda x: 'Unnamed' not in x)
monthly_CPI_non_seasonally_adjusted = pd.read_excel("Cleaned Data/CPI, Monthly, Non-Seasonally Adjusted.xlsx", usecols=lambda x: 'Unnamed' not in x)
monthly_inflation_rate_seasonally_adjusted = pd.read_excel("Cleaned Data/Inflation Rate, Monthly, Seasonally Adjusted.xlsx", usecols=lambda x: 'Unnamed' not in x)
monthly_inflation_rate_non_seasonally_adjusted = pd.read_excel("Cleaned Data/Inflation Rate, Monthly, Non-Seasonally Adjusted.xlsx" , usecols=lambda x: 'Unnamed' not in x)
monthly_houses_avg_price = pd.read_csv("Cleaned Data/Average House Price, Monthly.csv" , usecols=lambda x: 'Unnamed' not in x)
monthly_houses_sold = pd.read_csv("Cleaned Data/No. of Houses Sold, Monthly.csv" , usecols=lambda x: 'Unnamed' not in x)
```

```python
monthly_live_birth = monthly_live_birth.rename(columns={"data_series_monthly":"Year_Month"})
monthly_live_birth
```

```python
monthly_infant_death.columns.values[0] = "Year_Month"
monthly_infant_death
```

```python
monthly_marriage_number.rename(columns={monthly_marriage_number.columns[0]: 'Year_Month'},inplace=True)
monthly_marriage_number
```

```python
monthly_CPI_seasonally_adjusted.columns.values[0] = "Year_Month"
monthly_CPI_seasonally_adjusted
```

```python
monthly_CPI_non_seasonally_adjusted.columns.values[0] = "Year_Month"
monthly_CPI_non_seasonally_adjusted
```

```python
monthly_inflation_rate_seasonally_adjusted.columns.values[0] = "Year_Month"
monthly_inflation_rate_seasonally_adjusted.columns.values[1] = "Inflation(CPI % change from previous month, seasonally adjusted)"
monthly_inflation_rate_seasonally_adjusted
```

```python
monthly_inflation_rate_non_seasonally_adjusted.columns.values[0] = "Year_Month"
monthly_inflation_rate_non_seasonally_adjusted.columns.values[1] = "Inflation(CPI % change from previous month, non-seasonally adjusted)"
monthly_inflation_rate_non_seasonally_adjusted
```

```python
monthly_houses_avg_price.columns.values[0] = "Year_Month"
monthly_houses_avg_price.columns.values[1] = "Resale House Avg Price"
monthly_houses_avg_price["Year_Month"] = pd.to_datetime(monthly_houses_avg_price["Year_Month"])
monthly_houses_avg_price
```

```python
monthly_houses_sold.columns.values[0] = "Year_Month"
monthly_houses_sold.columns.values[1] = "Resale Houses Sold"
monthly_houses_sold["Year_Month"] = pd.to_datetime(monthly_houses_sold["Year_Month"])
monthly_houses_sold
```

```python
combined_data_together_overall = monthly_live_birth.copy()
combined_data_together_overall = combined_data_together_overall.merge(monthly_infant_death, on="Year_Month", how="left")
combined_data_together_overall = combined_data_together_overall.merge(monthly_marriage_number, on="Year_Month", how="left")
combined_data_together_overall = combined_data_together_overall.merge(monthly_CPI_seasonally_adjusted, on="Year_Month", how="left")
combined_data_together_overall = combined_data_together_overall.merge(monthly_CPI_non_seasonally_adjusted, on="Year_Month", how="left")
combined_data_together_overall = combined_data_together_overall.merge(monthly_inflation_rate_seasonally_adjusted, on="Year_Month", how="left")
combined_data_together_overall = combined_data_together_overall.merge(monthly_inflation_rate_non_seasonally_adjusted, on="Year_Month", how="left")

combined_data_together_overall = combined_data_together_overall.merge(monthly_houses_avg_price, on="Year_Month", how="left")
combined_data_together_overall = combined_data_together_overall.merge(monthly_houses_sold, on="Year_Month", how="left")
combined_data_together_overall
```

```python
combined_data_together_overall.to_excel("Cleaned Data/Monthly Data.xlsx")
```
