```python
# Basic Libraries
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
sb.set() # set the default Seaborn style for graphics
```

# LIVE BIRTHS? MOST IMPT? => 1960 to 2022

```python
live_birth_data = pd.read_excel("Raw Data/Live-Births By Sex And Ethnic Group, Monthly.xlsx", header = 9)
live_birth_data
```

```python
live_birth_no = live_birth_data.iloc[[0,1],1:] ##infant_death_no = infant_death_no.iloc[[0,1],1:]
live_birth_no
```

```python
live_birth_no_t = live_birth_no.T.copy()
live_birth_no_t.columns = ["data_series_monthly", "live births no."]
live_birth_no_t.index = list(range(756))
live_birth_no_t
```

```python
live_birth_no_t["data_series_monthly"] = pd.to_datetime(live_birth_no_t["data_series_monthly"])
live_birth_no_t.sort_values(by="data_series_monthly", inplace=True)
live_birth_no_t
```

```python
live_birth_no_t.plot(x = "data_series_monthly", y = "live births no.", figsize=(10, 6))
```

```python
live_birth_no_t.to_excel("Cleaned Data/Live Births.xlsx")
```
