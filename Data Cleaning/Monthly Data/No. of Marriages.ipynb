# NUMBER OF MARRIAGES PER MONTH? => 1961 to 2022


```python
marriages_data_raw = pd.read_excel("Raw Data/Marriages Under The Women's Charter By Ethnic Group Of Couple And Month Of Marriage, Annual.xlsx", header = 9)
marriages_data_raw = marriages_data_raw.iloc[0:14,:]
marriages_data_raw
```

```python
marriage_annual = marriages_data_raw.iloc[[0,1],1:]
marriage_annual_t = marriage_annual.T.copy()
marriage_annual_t.columns = ["Year", "no. of marriages"]
marriage_annual_t.index = list(range(62))
marriage_annual_t["Year"] = pd.to_datetime(marriage_annual_t["Year"], format = '%Y')
marriage_annual_t
marriage_annual_t.plot(x = "Year", y = "no. of marriages", figsize=(10, 6))
```
