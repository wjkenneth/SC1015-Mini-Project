```python
# Basic Libraries
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
sb.set() # set the default Seaborn style for graphics
```

```python
seasonally_adjusted_raw = pd.read_excel("Raw Data/Consumer Price Index (CPI), 2019 As Base Year, Monthly, Seasonally Adjusted.xlsx", header = 9)
non_adjusted_raw = pd.read_excel("Raw Data/Consumer Price Index (CPI), 2019 As Base Year, Monthly.xlsx", header = 9)
```

```python
seasonally_adjusted_raw
```

```python
seasonally_adjusted_overall = (seasonally_adjusted_raw.iloc[[0,1],1:].T).iloc[::-1].copy()
seasonally_adjusted_overall.index = list(range(seasonally_adjusted_overall.shape[0]))
seasonally_adjusted_overall.columns = ["Year_month", "CPI values(seasonally adjusted)"]
seasonally_adjusted_overall["Year_month"] = pd.to_datetime(seasonally_adjusted_overall["Year_month"])
seasonally_adjusted_overall["CPI values(seasonally adjusted)"] = seasonally_adjusted_overall["CPI values(seasonally adjusted)"].astype(float)
seasonally_adjusted_overall
```

```python
seasonally_adjusted_overall.plot(x="Year_month", y="CPI values(seasonally adjusted)", figsize=(10, 6))
```

```python
inflation_rate_seasonally_adjusted = pd.DataFrame(columns=["Year_month", "Inflation(change per month, from previous month, seasonally adjusted)"])
for i in range(1, seasonally_adjusted_overall.shape[0]):
    inflation_rate_seasonally_adjusted.loc[i] = [seasonally_adjusted_overall.iloc[i, 0], ((seasonally_adjusted_overall.iloc[i, 1] - seasonally_adjusted_overall.iloc[i-1, 1])/seasonally_adjusted_overall.iloc[i-1, 1])]
inflation_rate_seasonally_adjusted
```

```python
#inflation_rate_seasonally_adjusted.info()
inflation_rate_seasonally_adjusted.plot(x="Year_month", y="Inflation(change per month, from previous month, seasonally adjusted)", figsize=(10, 6))
```

```python
non_adjusted_raw_overall = (non_adjusted_raw.iloc[[0,1],1:].T).iloc[::-1].copy()
non_adjusted_raw_overall.index = list(range(non_adjusted_raw_overall.shape[0]))
non_adjusted_raw_overall.columns = ["Year_month", "CPI values(non-seasonally adjusted)"]
non_adjusted_raw_overall["Year_month"] = pd.to_datetime(non_adjusted_raw_overall["Year_month"])
non_adjusted_raw_overall["CPI values(non-seasonally adjusted)"] = non_adjusted_raw_overall["CPI values(non-seasonally adjusted)"].astype(float)
non_adjusted_raw_overall
```

```python
non_adjusted_raw_overall.plot(x="Year_month", y="CPI values(non-seasonally adjusted)", figsize=(10, 6))
```

```python
inflation_rate_non_seasonally_adjusted = pd.DataFrame(columns=["Year_month", "Inflation(change per month, from previous month, non-seasonally adjusted)"])
for i in range(1, non_adjusted_raw_overall.shape[0]):
    inflation_rate_non_seasonally_adjusted.loc[i] = [non_adjusted_raw_overall.iloc[i, 0], ((non_adjusted_raw_overall.iloc[i, 1] - non_adjusted_raw_overall.iloc[i-1, 1])/non_adjusted_raw_overall.iloc[i-1, 1])]
inflation_rate_non_seasonally_adjusted
```

```python
inflation_rate_non_seasonally_adjusted.plot(x="Year_month", y="Inflation(change per month, from previous month, non-seasonally adjusted)", figsize=(10, 6))
```

```python
seasonally_adjusted_overall.to_excel("Cleaned Data/CPI, Monthly, Seasonally Adjusted.xlsx")
inflation_rate_seasonally_adjusted.to_excel("Cleaned Data/Inflation Rate, Monthly, Seasonally Adjusted.xlsx")
non_adjusted_raw_overall.to_excel("Cleaned Data/CPI, Monthly, Non-Seasonally Adjusted.xlsx")
inflation_rate_non_seasonally_adjusted.to_excel("Cleaned Data/Inflation Rate, Monthly, Non-Seasonally Adjusted.xlsx")
```
