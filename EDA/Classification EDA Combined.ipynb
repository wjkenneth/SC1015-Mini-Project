```python
# Basic Libraries
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
sb.set() # set the default Seaborn style for graphics
```

```python
combined_annual_data = pd.read_excel("Combined Data.xlsx", usecols=lambda x: 'Unnamed' not in x)
```

```python
combined_annual_data
```

```python
temp_1 = combined_annual_data.copy()
temp_2 = pd.read_excel("Yearly Data.xlsx", usecols=lambda x: 'Unnamed' not in x).drop("number of live_birth in the year", axis=1)
temp_2
```

```python
combined_annual_data = temp_1.merge(temp_2, on="Year", how="left")
combined_annual_data
```

## Classification of Live Births (Increasing/Decreasing)

```python
annual_live_birth = combined_annual_data.iloc[:, :2]
annual_live_birth
```

```python
live_birth_annual_change = pd.DataFrame(columns=["Year_Month", "Change in Live Birth"])
for i in range(1, annual_live_birth.shape[0]):
    if ((annual_live_birth.iloc[i, 1] - annual_live_birth.iloc[i-1, 1]) > 0):
        change = "INCREASE"
    else:
        change = "DECREASE"
    live_birth_annual_change.loc[i] = [annual_live_birth.iloc[i, 0], change]   
live_birth_annual_change
```

## Converting to Change Dataset

```python
combined_changed_data_together_overall = pd.DataFrame(combined_annual_data.iloc[1:,0])
for i in range(1, combined_annual_data.shape[1]):    
    for j in range(1, combined_annual_data.shape[0]):
        combined_changed_data_together_overall.loc[j, i] = combined_annual_data.iloc[j, i] - combined_annual_data.iloc[j-1, i]
combined_changed_data_together_overall.columns = ["Year", "Change in Live Birth Number", "Change in Total Infant Death", "Change in number of marriages", "Change in CPI", "Change in Inflation", "Change in Resale House Avg Price", "Change in Number of Reslae Houses Sold", "Change in Total Population", "Change in Resident Population", "Change in Singapore Citizen Population", "Change in Permanent Resident Population","Change in Non-Resident Population","Change in Total Residents","Change in Residents Aged 20-39","Change in Female Residents Aged 20-39","Change in Total Brides Aged under 39","Change in Total Brides Aged under 44","Changes in Total Brides","Change in Overall Average Schooling Years","Change in Males' Average Schooling Years","Change in Females' Average Schooling Years","Change in Old-Age Support Ratio of Citizens Aged 20-64 per Citizen Aged 65 and over","Change in Old-Age Dependency Ratio of Citizens Aged 65 and over per Hundred Citizens Aged 20-64","Change in People with Minimum Education","Change in Females Aged 25-39  with Highest Education of Polytechnic / University","Change in Population with Highest Education of Polytechnic / University","Change in Median Income"," Change in Median Monthly Household Income","Change in Average Full-Time Working Hours per Week","Change in Fertility Rate","Change in Male Marriage Rate","Change in Female Marriage Rate","Change in Crude Marriage Rate","Change in Unemployment Rate"]
#combined_changed_data_together_overall = combined_changed_data_together_overall.merge(live_birth_monthly_change, on="Year_Month", how="left")
combined_changed_data_together_overall.insert(1, live_birth_annual_change.columns[1], live_birth_annual_change.iloc[:, 1])
combined_changed_data_together_overall
```

## Converting to % Change Dataset

```python
combined_changed_data_together_overall_percent = pd.DataFrame(combined_annual_data.iloc[1:,0])
for i in range(1, combined_annual_data.shape[1]):    
    for j in range(1, combined_annual_data.shape[0]):
        combined_changed_data_together_overall_percent.loc[j, i] = (combined_changed_data_together_overall.iloc[j-1, i+1])/(combined_annual_data.iloc[j-1, i])
combined_changed_data_together_overall_percent.columns = ["Year", "Change in Live Birth Number", "Change in Total Infant Death", "Change in number of marriages", "Change in CPI", "Change in Inflation", "Change in Resale House Avg Price", "Change in Number of Reslae Houses Sold", "Change in Total Population", "Change in Resident Population", "Change in Singapore Citizen Population", "Change in Permanent Resident Population","Change in Non-Resident Population","Change in Total Residents","Change in Residents Aged 20-39","Change in Female Residents Aged 20-39","Change in Total Brides Aged under 39","Change in Total Brides Aged under 44","Changes in Total Brides","Change in Overall Average Schooling Years","Change in Males' Average Schooling Years","Change in Females' Average Schooling Years","Change in Old-Age Support Ratio of Citizens Aged 20-64 per Citizen Aged 65 and over","Change in Old-Age Dependency Ratio of Citizens Aged 65 and over per Hundred Citizens Aged 20-64","Change in People with Minimum Education","Change in Females Aged 25-39  with Highest Education of Polytechnic / University","Change in Population with Highest Education of Polytechnic / University","Change in Median Income"," Change in Median Monthly Household Income","Change in Average Full-Time Working Hours per Week","Change in Fertility Rate","Change in Male Marriage Rate","Change in Female Marriage Rate","Change in Crude Marriage Rate","Change in Unemployment Rate"]
combined_changed_data_together_overall_percent.insert(1, live_birth_annual_change.columns[1], live_birth_annual_change.iloc[:, 1])
combined_changed_data_together_overall_percent 
```

## EDA on Raw Data

```python
combined_annual_data_temp = combined_annual_data.iloc[1:,:].copy()
combined_annual_data_temp.insert(1, live_birth_annual_change.columns[1], live_birth_annual_change.iloc[:, 1])
for col in combined_annual_data_temp.columns[2:]:
    f = plt.figure(figsize=(18, 6))
    sb.boxplot(x = col, y = "Change in Live Birth", data = combined_annual_data_temp, orient = "h")
```
