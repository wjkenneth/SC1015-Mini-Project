```python
# Basic Libraries
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
sb.set() # set the default Seaborn style for graphics
```

```python
combined_annual_data = pd.read_excel("Combined Data.xlsx", usecols=lambda x: 'Unnamed' not in x)
combined_annual_data
```

## Plotting the Data

```python
combined_annual_data.plot(x="Year", y=["number of live_birth in the year", "Total Infant Deaths", "number of marriages", "CPI values", "Inflation(CPI % change from previous year)", "Resale House Avg Price", "Resale Houses Sold"], figsize=(10, 6))
```

```python
combined_annual_data.describe()
```

## Normalising the Data for Easier Visualisation

```python
## normalising of data

## Normalise!!
from sklearn.preprocessing import MinMaxScaler
## Normalise!!
combined_annual_data_normalised = pd.DataFrame(combined_annual_data.iloc[:,0].copy())
for cols in range(1, combined_annual_data.shape[1]):
    values = combined_annual_data.iloc[:,cols].values
    values = values.reshape((len(values)), 1)
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaler = scaler.fit(values)
    normalized = scaler.transform(values)
    normalized_df = pd.DataFrame(normalized, columns=[combined_annual_data.columns.values[cols]])
    combined_annual_data_normalised = combined_annual_data_normalised.join(normalized_df)
combined_annual_data_normalised
```

```python
combined_annual_data_normalised.plot(x="Year", y=["number of live_birth in the year", "Total Infant Deaths", "number of marriages", "CPI values", "Inflation(CPI % change from previous year)", "Resale House Avg Price", "Resale Houses Sold"], figsize=(10, 6))
```

```python
combined_annual_data_normalised.plot(x="Year", y=["number of live_birth in the year", "Total Infant Deaths", "number of marriages"], figsize=(10, 6))
```

```python
combined_annual_data_normalised.plot(x="Year", y=["number of live_birth in the year", "CPI values", "Inflation(CPI % change from previous year)"], figsize=(10, 6))
```

```python
combined_annual_data_normalised.plot(x="Year", y=["number of live_birth in the year", "Resale House Avg Price", "Resale Houses Sold"], figsize=(10, 6))
```

## Standardising the Data for Easier Visualisation

```python
## Standardise!!
from sklearn.preprocessing import StandardScaler

## Standardise!!
combined_annual_data_standardised = pd.DataFrame(combined_annual_data.iloc[:,0].copy())
for cols in range(1, combined_annual_data.shape[1]):
    values = combined_annual_data.iloc[:,cols].values
    values = values.reshape((len(values)), 1)
    scaler = StandardScaler()
    scaler = scaler.fit(values)
    standardised = scaler.transform(values)
    standardised_df = pd.DataFrame(standardised, columns=[combined_annual_data.columns.values[cols]])
    combined_annual_data_standardised = combined_annual_data_standardised.join(standardised_df)
#combined_annual_data
combined_annual_data_standardised
```

```python
combined_annual_data_standardised.plot(x="Year", y=["number of live_birth in the year", "Total Infant Deaths", "number of marriages", "CPI values", "Inflation(CPI % change from previous year)", "Resale House Avg Price", "Resale Houses Sold"], figsize=(10, 6))
```

```python
combined_annual_data_standardised.plot(x="Year", y=["number of live_birth in the year", "Total Infant Deaths", "number of marriages"], figsize=(10, 6))
```

```python
combined_annual_data_standardised.plot(x="Year", y=["number of live_birth in the year", "CPI values", "Inflation(CPI % change from previous year)"], figsize=(10, 6))
```

```python
combined_annual_data_standardised.plot(x="Year", y=["number of live_birth in the year", "Resale House Avg Price", "Resale Houses Sold"], figsize=(10, 6))
```
